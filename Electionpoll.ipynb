```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\election2024.csv")

# Display the first few rows of the dataset
df.head()
```

```python
# Display the last few rows of the dataset
df.tail()
```

```python
# Get a summary of the dataset
df.info()
```

```python
# Get descriptive statistics
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

Data Cleaning

```python
# Handling missing values
df.dropna(inplace=True)
```

```python
# Removing duplicates 
df.drop_duplicates(inplace=True)
```

```python
df['job_type'].fillna(df['company_type'].mode()[0], inplace=True)
df['company_type'].fillna(df['company_type'].mode()[0], inplace=True)
 
```

```python

df.isnull().sum()
```

Data Visualization

```python
df.columns
```

```python
# Distribution of age
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Poll results by region
plt.figure(figsize=(10, 6))
sns.barplot(x='will_vote', y='age', data=df, estimator=np.mean)
plt.title('Who will vote according to age')
plt.xlabel('will_vote')
plt.ylabel('Age')
plt.show()
```

```python
df.columns
```

```python
# Candidate preference
plt.figure(figsize=(10, 6))
sns.countplot(x='candidate', data=df)
plt.title('Candidate Preference')
plt.xlabel('Candidate')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Voting intention
plt.figure(figsize=(10, 6))
sns.countplot(x='will_vote', data=df)
plt.title('Voting Intention')
plt.xlabel('Will Vote')
plt.ylabel('Count')
plt.show()
```

```python
# Regional distribution of votes
plt.figure(figsize=(12, 8))
sns.countplot(x='federal_district', hue='candidate', data=df)
plt.title('Regional Distribution of Votes')
plt.xlabel('Federal District')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top candidates by count
top_candidates = df['candidate'].value_counts()
print(top_candidates)
```

```python
# Regional analysis
regional_analysis = df.groupby('federal_district')['candidate'].value_counts().unstack().fillna(0)
print(regional_analysis)
```
