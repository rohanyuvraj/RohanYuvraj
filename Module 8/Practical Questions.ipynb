Protected attributes are intended to be accessible only within the class and its subclasses. By convention, protected attributes are prefixed with a single underscore (_). This is a convention and not enforced by Python.

````python
class BaseClass:
    def __init__(self):
        self._protected_attribute = "I am a protected attribute"

class DerivedClass(BaseClass):
    def print_protected(self):
        print(self._protected_attribute)

base = BaseClass()
print(base._protected_attribute) 

derived = DerivedClass()
derived.print_protected()  
````

Private attributes are intended to be accessible only within the class they are defined in. They are prefixed with a double underscore (__). 
Python performs name mangling on these attributes to make them harder to access from outside the class.

````python
class MyClass:
    def __init__(self):
        self.__private_attribute = "I am a private attribute"

    def get_private_attribute(self):
        return self.__private_attribute

my_object = MyClass()


# Accessing private attribute using a public method
print(my_object.get_private_attribute())  

# Accessing private attribute with name mangling
print(my_object._MyClass__private_attribute)  
````

A decorator is essentially a function that takes another function (or method) as an argument and returns a new function that typically extends or alters the behavior of the original function.

````python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
````

Generators are a special type of iterator created using functions with the yield keyword. They provide a convenient way to create iterators without having to implement the iterator protocol manually.

````python
# Define a generator function
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

# Create a generator object
counter = count_up_to(5)

# Iterate over the generator
for number in counter:
    print(number)
````

````python
def calculator():
    print("Welcome to the Simple Calculator!")
    print("Select operation:")
    print("1. Addition (+)")
    print("2. Subtraction (-)")
    print("3. Multiplication (*)")
    print("4. Division (/)")

    # Take input from the user
    operation = input("Enter operation (1/2/3/4): ")
    
    # Check if the operation is valid
    if operation in ['1', '2', '3', '4']:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if operation == '1':
            result = num1 + num2
            print(f"{num1} + {num2} = {result}")
        elif operation == '2':
            result = num1 - num2
            print(f"{num1} - {num2} = {result}")
        elif operation == '3':
            result = num1 * num2
            print(f"{num1} * {num2} = {result}")
        elif operation == '4':
            if num2 != 0:
                result = num1 / num2
                print(f"{num1} / {num2} = {result}")
            else:
                print("Error: Division by zero is not allowed.")
    else:
        print("Invalid operation selected.")

# Run the calculator
calculator()
````

````python
# Function to check if a number is prime
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# List to store prime numbers
prime_numbers = []

# Loop through numbers from 2 to 100
for number in range(2, 101):
    if is_prime(number):
        prime_numbers.append(number)

print("Prime numbers between 2 and 100 are:", prime_numbers)
````

````python
def main():
    # a) Accept the marks of the student in five major subjects
    marks = []
    subjects = ["Subject 1"" "Subject 2", "Subject 3", "Subject 4", "Subject 5"]
    
    for subject in subjects:
        mark = float(input(f"Enter marks for {subject}: "))
        marks.append(mark)

    # b) Display the marks and calculate total and percentage
    print("\nMarks entered:")
    for i, mark in enumerate(marks, start=1):
        print(f"{subjects[i-1]}: {mark}")

    total_marks = sum(marks)
    percentage = total_marks / 5

    print(f"\nTotal Marks: {total_marks}")
    print(f"Percentage: {percentage:.2f}%")

    # c) Find the grade of the student using match-case
    match percentage:
        case p if p > 85:
            grade = 'A'
        case p if 75 <= p <= 85:
            grade = 'B'
        case p if 50 <= p < 75:
            grade = 'C'
        case p if 30 <= p < 50:
            grade = 'D'
        case _:
            grade = 'Reappear'

    print(f"Grade: {grade}")

if __name__ == "__main__":
    main()
````

````python
import math

# Constants
G = 6.67430e-11  # Gravitational constant in m^3 kg^-1 s^-2
mass_earth = 5.972624e24  # Mass of Earth in kilograms
mass_moon = 7.34767309e22  # Mass of Moon in kilograms
mass_sun = 1.989630e30  # Mass of Sun in kilograms
distance_earth_sun = 1496e11  # Average distance between Earth and Sun in meters
distance_moon_earth = 3844e8  # Average distance between Moon and Earth in meters

def calculate_gravitational_force(mass1, mass2, distance):
    """
    Calculate the gravitational force between two objects using the formula:
    F = G \* (m1 \* m2) / r^2
    
    Args:
        mass1 (float): Mass of the first object in kilograms.
        mass2 (float): Mass of the second object in kilograms.
        distance (float): Distance between the two objects in meters.
    
    Returns:
        float: Gravitational force between the two objects in Newtons.
    """
    return G * (mass1 * mass2) / (distance ** 2)

# Calculate gravitational forces
force_earth_sun = calculate_gravitational_force(mass_earth, mass_sun, distance_earth_sun)
force_moon_earth = calculate_gravitational_force(mass_moon, mass_earth, distance_moon_earth)

# Compare forces to determine which one is stronger
if force_earth_sun > force_moon_earth:
    print("The gravitational force between the Earth and the Sun is stronger.")
else:
    print("The gravitational force between the Moon and the Earth is stronger.")

# Determine which celestial body is more attracted to the other
if force_moon_earth > force_earth_sun:
    print("The Moon is more attracted to the Earth.")
else:
    print("The Earth is more attracted to the Sun.")

print(f"Gravitational force between Earth and Sun: {force_earth_sun:.2e} N")
print(f"Gravitational force between Moon and Earth: {force_moon_earth:.2e} N")
````

````python
class Student:
    def __init__(self, name, age, roll_number):
        self.__name = name
        self.__age = age
        self.__roll_number = roll_number

    # Getter methods
    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    def get_roll_number(self):
        return self.__roll_number

    # Setter methods
    def set_name(self, name):
        self.__name = name

    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Invalid age. Please enter a positive value.")

    def set_roll_number(self, roll_number):
        if roll_number > 0:
            self.__roll_number = roll_number
        else:
            print("Invalid roll number. Please enter a positive value.")

    # Method to display student information
    def display_info(self):
        print(f"Name: {self.get_name()}")
        print(f"Age: {self.get_age()}")
        print(f"Roll Number: {self.get_roll_number()}")

    # Method to update student details
    def update_details(self, name=None, age=None, roll_number=None):
        if name is not None:
            self.set_name(name)
        if age is not None:
            self.set_age(age)
        if roll_number is not None:
            self.set_roll_number(roll_number)

# Create instances of the "Student" class and test the implemented functionality
student1 = Student("John Doe", 20, 101)
student2 = Student("Jane Smith", 22, 102)

print("Initial Student Information:")
student1.display_info()
student2.display_info()

# Update student details
student1.update_details(name="John D.", age=21)
student2.update_details(roll_number=103)

print("\nUpdated Student Information:")
student1.display_info()
student2.display_info()
````

````python
class LibraryBook:
    def __init__(self, book_name, author):
        self.__book_name = book_name
        self.__author = author
        self.__is_available = True

    # Getter methods
    def get_book_name(self):
        return self.__book_name

    def get_author(self):
        return self.__author

    def is_available(self):
        return self.__is_available

    # Setter methods (not directly accessible to ensure encapsulation)
    def __set_availability(self, status):
        self.__is_available = status

    # Method to borrow a book
    def borrow_book(self):
        if self.is_available():
            print(f"You have borrowed '{self.get_book_name()}' by {self.get_author()}.")
            self.__set_availability(False)
        else:
            print(f"'{self.get_book_name()}' by {self.get_author()} is currently not available.")

    # Method to return a book
    def return_book(self):
        if not self.is_available():
            print(f"You have returned '{self.get_book_name()}' by {self.get_author()}.")
            self.__set_availability(True)
        else:
            print(f"'{self.get_book_name()}' by {self.get_author()} is already available.")

# Test the borrowing and returning functionality with sample data
book1 = LibraryBook("To Kill a Mockingbird", "Harper Lee")
book2 = LibraryBook("1984", "George Orwell")

print("Initial Book Availability:")
print(f"{book1.get_book_name()} by {book1.get_author()}: {'Available' if book1.is_available() else 'Not Available'}")
print(f"{book2.get_book_name()} by {book2.get_author()}: {'Available' if book2.is_available() else 'Not Available'}")

# Borrow books
book1.borrow_book()
book2.borrow_book()

print("\nBook Availability after Borrowing:")
print(f"{book1.get_book_name()} by {book1.get_author()}: {'Available' if book1.is_available() else 'Not Available'}")
print(f"{book2.get_book_name()} by {book2.get_author()}: {'Available' if book2.is_available() else 'Not Available'}")

# Return books
book1.return_book()
book2.return_book()

print("\nBook Availability after Returning:")
print(f"{book1.get_book_name()} by {book1.get_author()}: {'Available' if book1.is_available() else 'Not Available'}")
print(f"{book2.get_book_name()} by {book2.get_author()}: {'Available' if book2.is_available() else 'Not Available'}")
````

````python
class MenuItem:
    _next_id = 1
    _menu = {}

    def __init__(self, name, description, price, category):
        self._id = MenuItem._next_id
        self.name = name
        self.description = description
        self.price = price
        self.category = category
        MenuItem._menu[self._id] = self
        MenuItem._next_id += 1

    @classmethod
    def add_menu_item(cls, name, description, price, category):
        new_item = cls(name, description, price, category)
        return new_item

    @classmethod
    def update_menu_item(cls, id, attribute_name, new_value):
        if id in cls._menu:
            setattr(cls._menu[id], attribute_name, new_value)
            print(f"Updated {attribute_name} of menu item {id} to {new_value}")
        else:
            print("Menu item not found")

    @classmethod
    def remove_menu_item(cls, id):
        if id in cls._menu:
            del cls._menu[id]
            print(f"Removed menu item {id}")
        else:
            print("Menu item not found")

    @property
    def id(self):
        return self._id
````

Here are the steps to create a Flask Blueprint and why you might use one:

**What is a Flask Blueprint?**

A Flask Blueprint is a way to organize and structure your Flask application into smaller, reusable components.

**Why use a Flask Blueprint?**

You might use a Flask Blueprint to:

* Organize your application into logical components (e.g., user authentication, API endpoints)
* Reuse code across multiple applications
* Simplify large applications by breaking them down into smaller pieces

**Steps to create a Flask Blueprint:**

1. **Create a new Python file**: Create a new Python file for your blueprint (e.g., `user_blueprint.py`)
2. **Import Flask and Blueprint**: Import the `Flask` and `Blueprint` classes from the `flask` module
3. **Create a new Blueprint instance**: Create a new instance of the `Blueprint` class, passing in the name of your blueprint (e.g., `user_blueprint`)
4. **Define routes and views**: Define routes and views for your blueprint using the `@blueprint.route()` decorator
5. **Register the blueprint with your app**: Register your blueprint with your main Flask application using the `app.register_blueprint()` method

Example:
```python
# user_blueprint.py
from flask import Blueprint, render_template

user_blueprint = Blueprint('user', __name__)

@user_blueprint.route('/login')
def login():
    return render_template('login.html')

# app.py
from flask import Flask
from user_blueprint import user_blueprint

app = Flask(__name__)
app.register_blueprint(user_blueprint)
```
By using Blueprints, you can keep your application organized, modular, and easy to maintain!

````python
**Series:**

A `Series` is a one-dimensional labeled array of values. It's similar to a list, but with additional features like labels (index) and support for missing data. Think of it as a column of data with row labels.

Example:
```python
import pandas as pd

# create a series
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print(s)
```
Output:
```
a    1
b    2
c    3
d    4
e    5
dtype: int64
```
**DataFrame:**

A `DataFrame` is a two-dimensional labeled data structure with columns of potentially different types. It's similar to an Excel spreadsheet or a table in a relational database. Think of it as a collection of Series objects with shared index (row labels).

Example:
```python
import pandas as pd

# create a dataframe
data = {'Name': ['John', 'Mary', 'David'],
        'Age': [25, 31, 42]}
df = pd.DataFrame(data)
print(df)
```
Output:
```
     Name  Age
0     John   25
1     Mary   31
2    David   42
```
Key differences:

1. **Dimensionality**: Series is one-dimensional (a single column), while DataFrame is two-dimensional (multiple columns).
2. **Structure**: Series has only one set of labels (index), while DataFrame has both row labels (index) and column labels.
3. **Data types**: Series can hold only one type of data, while DataFrame can hold different types of data in each column.

When to use each:

* Use Series when working with a single column of data or when you need to perform operations on individual columns.
* Use DataFrame when working with multiple columns or when you need to perform operations on entire rows or subsets of rows.

````

````python


In pandas, `loc` and `iloc` are two ways to access rows and columns in a DataFrame. The main difference between them is how they handle indexing:

**loc (Label-based indexing)**

* `loc` uses label-based indexing, which means you specify rows and columns based on their labels (i.e., the values in the index).
* It is primarily label-based, but may also be used with a boolean array.
* It allows you to access rows and columns by their label names.

Example:
```python
import pandas as pd

# create a sample dataframe
data = {'Name': ['John', 'Mary', 'David'],
        'Age': [25, 31, 42]}
df = pd.DataFrame(data)

# access row with label '1' (second row)
print(df.loc[1])

# access column with label 'Name'
print(df.loc[:, 'Name'])
```
Output:
```
Name    Mary
Age      31
Name      David
Age        42

0      John
1      Mary
2     David
```
**iloc (Integer-based indexing)**

* `iloc` uses integer-based indexing, which means you specify rows and columns by their integer position (i.e., the numerical index).
* It is primarily integer position-based.
* It allows you to access rows and columns by their integer position.

Example:
```python
import pandas as pd

# create a sample dataframe
data = {'Name': ['John', 'Mary', 'David'],
        'Age': [25, 31, 42]}
df = pd.DataFrame(data)

# access second row (integer position 1)
print(df.iloc[1])

# access first column (integer position 0)
print(df.iloc[:, 0])
```
Output:
```
Name    Mary
Age      31

0      John
1      Mary
2     David
```
Key differences:

* **Indexing style**: `loc` uses label-based indexing, while `iloc` uses integer-based indexing.
* **Row/column selection**: With `loc`, you can select rows and columns by their labels. With `iloc`, you can select rows and columns by their integer position.

When to use each:

* Use `loc` when working with DataFrames that have meaningful index labels (e.g., dates, names) or when you need to perform operations based on
````
